```python
class Characteristic_function:
    def __init__(self,data):
        self.data = data
    def ecf(self,t):
        n = len(self.data)
        if type(t)== int:
            C = np.zeros((n,1))
            S = np.zeros((n,1))
            for i in range(n):
                C[i] = math.cos(t*self.data[i])
                S[i] = math.sin(t*self.data[i])
            s = np.mean(S)
            c = np.mean(C)
            return c,s                
        else:    
            m = len(t)
            C = np.zeros((n,1))
            S = np.zeros((n,1))
            R = np.zeros((m,2))
            for j in range(m):
                for i in range(n):
                    C[i] = math.cos(t[j]*self.data[i])
                    S[i] = math.sin(t[j]*self.data[i])
                s = np.mean(S)
                c = np.mean(C)
                R[j,0] = c
                R[j,1] = s
        return R
            
            
```

```python
def cumulate(data):
    delta = [0]
    n = len(data)
    for i in range(n):
        delta.append(delta[-1]+data[i])
    return delta
```

```python
df = 3
size = (1000,1)
data = np.random.standard_t(df, size)

plt.plot(data)
```

```python
delta = cumulate(data)
plt.plot(delta)
```

```python
ECF = Characteristic_function(data)
ECF.ecf(2)
```

```python
M = 50
N = 100
feature = np.linspace(-M, M, num=N, endpoint=True, retstep=False, dtype=None, axis=0)
labels = ECF.ecf(feature)
```
